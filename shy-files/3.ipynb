运动员个人的总奖牌数

```python
from statics.Statics import Statics
import plotly.express as px  
import plotly.graph_objects as go  
import pandas as pd  

self = Statics()

df = self.athlete.csv_file

df.drop_duplicates(subset=['NOC', 'Sport', 'Event', 'Medal'], inplace=True)

df = df.groupby(['Name']).size().reset_index(name='Medals')

# df = df.groupby(['Medals']).size().reset_index(name='Number')
# df.to_csv('medals.csv', index=False)

df = df.sort_values(by=['Medals'], ascending=False)

df.to_csv('medals.csv', index=False)

# 绘制频率直方图  
fig = go.Figure()  

# 添加直方图  
fig.add_trace(go.Histogram(  
    x=df['Medals'],  # 奖牌数量  
    nbinsx=10,  # 直方图的分箱数量，可以根据需要调整  
    marker=dict(  
        color='blue',  # 块的颜色  
        line=dict(  
            color='black',  # 分界线颜色  
            width=1.5       # 分界线宽度  
        )  
    )  
))  

# 设置图表标题和轴标签  
fig.update_layout(  
    title='运动员奖牌数量频率直方图',  
    xaxis_title='奖牌数量',  
    yaxis_title='频率',  
    font=dict(  
        family="Arial, sans-serif",  # 字体  
        size=14,                     # 字体大小  
        color="black"                # 字体颜色  
    ),  
    bargap=0.2,  # 设置柱状图块之间的间距  
    template='plotly_white',  # 使用专业的白色背景模板  
    width=800,  # 图表宽度  
    height=600  # 图表高度  
)  

# 保存为图片（适合论文）  
# fig.write_image("histogram_plot.png", format="png", scale=3)  # scale=3 提高分辨率  

# 显示图表  
fig.show()  
```
